## Comparador de Preço Oeste Trigo

```python
import pandas as pd
```

##### Importação do Csv

```python
colunas=['CodProduto','Produto','PreçoAtual','PreçoAntigo']
data = pd.read_csv("210619.csv",na_values='.')
data.columns=colunas
data.head()
```

##### Criação das novas colunas

```python
data['Diferença']=0.0
data['Porcentagem']=0.0
data['Status']='Null'
```

```python
data.head()
```

##### Classificação dos preços

```python
for i in range (0,data.CodProduto.count()):
    if data.PreçoAntigo[i] < data.PreçoAtual[i] and data.PreçoAntigo[i]>0 :
        porcentagem=((100*data.PreçoAtual[i])/data.PreçoAntigo[i])-100
        if porcentagem >=3:
            data['Status'][i]="Mais caro"
            diferencaPreco = data.PreçoAtual[i] - data.PreçoAntigo[i]
            data['Diferença'][i]=diferencaPreco
            data['Porcentagem'][i]=porcentagem
    else:
        porcentagem=((100*data.PreçoAntigo[i])/data.PreçoAtual[i])-100
        if porcentagem >3:
            data['Status'][i]="Mais Barato"
            data['Diferença'][i]= data.PreçoAntigo[i] - data.PreçoAtual[i]
            data['Porcentagem'][i]=porcentagem
        
```

```python
data[data['Status']=='Mais Barato']
```

##### Criação do excel para enviar para a empresa  com os preços mais baixo
